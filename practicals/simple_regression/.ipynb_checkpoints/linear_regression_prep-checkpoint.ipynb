## Importance sampling

```python
class ImportanceSampling():
    
    def __init__(self, dimension, log_target, num_samples):
        """initialize class
        """
        self.dimension = dimension
        self.log_target = log_target
        self.num_samples = num_samples
        return
    
    def find_map(self):
        """find MAP to locate proposal there and scale it using the inverse Hessian
        """
        res = spo.minimize(lambda theta: -self.log_target(theta), np.zeros((self.dimension,)))
        self.map = res['x']
        self.hess_inv = res['hess_inv']    
        return
    
    def propose(self):
        proposal = sps.multivariate_normal(mean=self.map, cov=5*self.hess_inv)
        self.nodes = proposal.rvs(size=self.num_samples)
        unnormalized_log_weights = np.array([self.log_target(node) - proposal.logpdf(node) for node in self.nodes])
        self.log_weights = unnormalized_log_weights - logsumexp(unnormalized_log_weights)
        self.weights = np.exp(self.log_weights) # weights are normalized
        return
    
    def get_estimate(self, f):
        """estimate the integral of function f wrt the target
        """
        self.function_values = np.array([f(node) for node in self.nodes])
        self.estimate = np.sum([w*f for w, f in zip(self.weights, self.function_values)], 0)
        return self.estimate
    
    def get_sample(self, num_resamples):
        """sample from the IS approximation
        """
        indices = npr.choice(np.arange(self.num_samples), size=num_resamples, p=self.weights)
        return [self.nodes[i] for i in indices]        

    def get_ess(self):
        """compute and return effetive sample size
        """
        N = self.num_samples
        variance_estimate = 1/(N-1)*np.sum(IS.weights-1/N)**2
        self.ess = N/(1 + variance_estimate)
        return self.ess
```

```python
# Go hierarchical

## Fetch data
X = dict()
y = dict()
X["African"] = african_nations["rugged"].values
y["African"] = african_nations["rgdppc_2000"].values
X["Rest"] = non_african_nations["rugged"].values
y["Rest"] = non_african_nations["rgdppc_2000"].values

n = len(X)
d = 1
num_resamples = 1000

def log_prior(theta):
    prior_mean = np.array([np.mean(y["African"]), np.exp(theta[6]), 0, 
                           np.mean(y["Rest"]), np.exp(theta[6]), 0])
    return -npl.norm(theta[:6]-prior_mean)**2/2 - theta[6]**2/2 # theta = (b, a, log_sigma)

def log_likelihood(theta):
    return (
        - theta[2] - npl.norm(y["African"] - X["African"]*theta[1]-theta[0])**2 / (2*np.exp(theta[2])**2 )
        - theta[5] - npl.norm(y["Rest"] - X["Rest"]*theta[4]-theta[3])**2 / (2*np.exp(theta[5])**2 )
    )
    
unnormalized_log_posterior = lambda theta: log_prior(theta) + log_likelihood(theta)

## Perform IS
IS = ImportanceSampling(7, unnormalized_log_posterior, num_samples=10000)
IS.find_map()
print("MAP is", IS.map)
print("Diag of inv. Hessian is", np.diag(IS.hess_inv))
IS.propose()
samples = IS.get_sample(num_resamples)

## plot
```

```python
for label in ["African", "Rest"]:
    sns.scatterplot(X[label], y[label])
    offset = 0 + 3*(label=="Rest")
    for theta in samples:
        plt.plot(X[label], theta[offset] + theta[1+offset]*X[label], color="orange", alpha=.1)
        plt.title(label)
    plt.show()
```
